```python
train_data = dataset["train"]
train_df = pd.DataFrame(train_data)
validation_data=dataset["validation"]
validation_df=pd.DataFrame(validation_data)
test_data=dataset["test"]
test_df=pd.DataFrame(test_data)
```

```python
pos=[]
chunks=[]
ner=[]
tokens=[]
firsts=[]
for i in range(train_df.shape[0]):
    pos.extend(train_df.iloc[i]["pos_tags"])
    tokens.extend(train_df.iloc[i]["tokens"])
    chunks.extend(train_df.iloc[i]["chunk_tags"])
    ner.extend(train_df.iloc[i]["ner_tags"])
X_train=pd.DataFrame(columns=['POS','Chunks'])
y_train=pd.DataFrame(columns=['NER'])

X_train['POS']=pos
X_train['Chunks']=chunks

digits=[]
case=[]
title=[]
for i in tokens:
    if i.isdigit()==True:
            digits.append(1)
            case.append(0)
            title.append(0)
    else:
        digits.append(0)
        if i[0].isupper():
            case.append(1)
        else:
            case.append(0)
        if i.istitle():
            title.append(1)
        else:
            title.append(0)

X_train['Case']=case
X_train['Digits']=digits
X_train['Title']=title

mod=[]
for i in ner:
    if i==0:
        mod.append(0)
    else:
        mod.append(1)
y_train['NER']=mod
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import numpy as np

data=train_df

# Convert tokens, pos_tags, and chunk_tags to strings
tokens_str = [" ".join(sentence) for sentence in data["tokens"]]
pos_tags_str = [" ".join(map(str, tags)) for tags in data["pos_tags"]]
chunk_tags_str = [" ".join(map(str, tags)) for tags in data["chunk_tags"]]

# Combine features into a feature matrix
features = [f"{tokens} {pos_tags} {chunk_tags}" for tokens, pos_tags, chunk_tags in zip(tokens_str, pos_tags_str, chunk_tags_str)]

# Convert NER tags to numerical labels
mlb = MultiLabelBinarizer()
labels = mlb.fit_transform(data["ner_tags"])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Vectorize the text data
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)


y_train_flat = np.ravel(y_train)
y_test_flat = np.ravel(y_test)

# Train an SVM classifier
svm_classifier = SVC(kernel="linear")
svm_classifier.fit(X_train, y_train)

# Predict on the test data
predictions = svm_classifier.predict(X_test_vec)

# Convert numerical predictions back to NER tags
predicted_labels = mlb.inverse_transform(predictions)

# Print classification report
print(classification_report(y_test_flat, predictions, target_names=mlb.classes_))
```

```python
print("X_train_vec shape:", X_train_vec.shape)
print("X_test_vec shape:", X_test_vec.shape)
print("y_train_flat shape:", y_train.shape)
print("y_test_flat shape:", y_test.shape)
```
